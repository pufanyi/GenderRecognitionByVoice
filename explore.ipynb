```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

sns.set(font_scale=1.5)

data = pd.read_csv('data/cleaned/voice_data.csv')
X_train = pd.read_csv('data/cleaned/X_train.csv')
y_train = pd.read_csv('data/cleaned/y_train.csv')
X_test = pd.read_csv('data/cleaned/X_test.csv')
y_test = pd.read_csv('data/cleaned/y_test.csv')
```

```python
from sklearn.svm import SVC

svc = SVC(random_state=87).fit(X_train, y_train.values.ravel())
svc.score(X_test, y_test)
```

```python
y_train_predict = svc.predict(X_train)
y_test_predict = svc.predict(X_test)

X_test_false = X_test[y_test_predict != y_test.values.ravel()]
y_test_false = y_test[y_test_predict != y_test.values.ravel()]
X_train_false = X_train[y_train_predict != y_train.values.ravel()]
y_train_false = y_train[y_train_predict != y_train.values.ravel()]
```

```python
pd.concat([X_test_false, y_test_false], axis=1).to_csv('data/result/test_false.csv', index=False)
pd.concat([X_train_false, y_train_false], axis=1).to_csv('data/result/train_false.csv', index=False)
```

Automatically choose the best model for svm

```python
from sklearn.model_selection import GridSearchCV

param_grid = [
    {
        'kernel': ['linear'],
        'C': [0.1, 1, 10, 100, 1000]
    },
    {
        'kernel': ['rbf'],
        'C': [0.1, 1, 10, 100, 1000],
        'gamma': [0.001, 0.0001]
    }
]

svc = SVC(random_state=87)
grid_search = GridSearchCV(svc, param_grid, cv=5, scoring='accuracy', return_train_score=True)
grid_search.fit(X_train, y_train.values.ravel())
```
